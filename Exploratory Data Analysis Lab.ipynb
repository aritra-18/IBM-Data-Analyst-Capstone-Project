## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
df_cc= pd.DataFrame(data= df['ConvertedComp'])
df_cc.plot(kind='box', figsize= (10,8))
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
Q1,Q3 = df['ConvertedComp'].quantile(0.25),df['ConvertedComp'].quantile(0.75)
IQR = Q3-Q1
print("Inter Quartile Range for the column ConvertedComp is ", IQR)
```

Find out the upper and lower bounds.


```python
# your code goes here
upper_bound = Q3+(IQR*1.5)
lower_bound = Q1-(IQR*1.5)

print("upper bound is", upper_bound)
print("lower bound is", lower_bound)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
outliers=(df['ConvertedComp'] < lower_bound) & (df['ConvertedComp'] > upper_bound)
len(outliers)
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
df_filtered = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]
df_no_outliers = pd.DataFrame(df_filtered)
df_no_outliers.describe()
```

```python
#median after removing outliers
df_no_outliers['ConvertedComp'].median()
```

```python
#mean after removing outliers
df_no_outliers['ConvertedComp'].mean()
```
